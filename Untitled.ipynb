```python
import pandas as pd
df=pd.read_excel("dataCleaning.xlsx", sheet_name="crosstab")
```

```python
df
```

```python
# Crosstabulation without row or column totals
pd.crosstab(df["SOCIALMEDIA"], df["GENDER"])
```

```python
# Crosstabulation with row and column totals
pd.crosstab(df["SOCIALMEDIA"], df["GENDER"], margins=True)
```
